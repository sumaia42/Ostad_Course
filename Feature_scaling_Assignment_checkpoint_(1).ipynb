### Read data from income.csv and scale the relevant features using standardization

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/income.csv")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
X = df[["Age", "Income"]]
y = df["Buy"]
```

```python
scaler = StandardScaler()
```

```python
scale1 = scaler.fit_transform(X)
```

```python
scaleddf = pd.DataFrame(scale1, columns=["Age", "Income"])
```

```python
final = pd.concat([scaleddf,y],axis=1)
```
